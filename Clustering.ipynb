```python
#First start the Carrot2 clustering engine
#$cd carrot2-dcs-3.9.0/ 
#$./dcs.sh 

#TODO log
import requests
from xml.sax.saxutils import escape
from collections import OrderedDict
from collections import defaultdict
   
class Clusters(OrderedDict):
    @classmethod
    def convert_docs_to_xml(cls,users,title_field="title",snippet_field="snippet",url_field="url"):
        xml = ["<searchresult>"]
        for user in users :
            if snippet_field not in user:
                continue
            xml.append("<document>")
            xml.append("<title>")
            xml.append(escape(user[title_field]))
            xml.append("</title>")
            xml.append("<snippet>")
            xml.append(escape(user[snippet_field]))
            xml.append("</snippet>")
            xml.append("<url>")
            if url_field in user:
                xml.append(escape(user[url_field]))
            xml.append("</url>")
            xml.append("</document>")        
        xml.append("</searchresult>")
        return "".join(xml)

    def __init__(self,hydrated_users,title_field="name",snippet_field="description",url_field="url"):
        url = "http://localhost:8080/dcs/rest" #TODO externalize this
        payload = {
            "dcs.c2stream": Clusters.convert_docs_to_xml(hydrated_users,title_field=title_field,snippet_field=snippet_field,url_field=url_field),
            "dcs.output.format": "JSON"
        }
        r = requests.post(url, data=payload)
        #TODO error test
        j = r.json()
        clusters = defaultdict(list)
        for c in j["clusters"]:
            clusters[c["size"]].extend(c["phrases"])
        
        super(Clusters,self).__init__(reversed(sorted(clusters.items())))
        #after constructor runs make sure that no new users can be added
        self.__setattr__ = self._unimplemented

    def _unimplemented(self,key,val):
        raise Exception("not supported")
```
